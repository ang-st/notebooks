```python
def cci(d, cci_win=16, meanwin=16):
    a=pd.DataFrame(index=d.index)
    a["cci"]=CCI(d,cci_win)
    a["cci_smooth"]=pd.rolling_mean(a.cci,meanwin)
    tw=pd.DataFrame(index=a.index)
    tw["signal"]=0
    tw.loc[a.cci > a.cci_smooth,"signal"]=1
    tw.loc[a.cci < a.cci_smooth,"signal"]=0.#-1
    tw["returns"]=d.close.pct_change().shift(-1)
    return tw
```

```python
def ma_cross(data, win_short=50, win_long=208):

    #long_sma  = pd.rolling_mean(data.price, win_short)
    lreg_long=pd.rolling_mean(data.close,win_long)
    lreg_short=pd.rolling_mean(data.close,win_short)
    # target weights
    tw = pd.DataFrame(index=data.index.copy(),columns=["signal"])
    #tw = long_sma.copy()
    #tw[:]=0
    tw.loc[:,"signal"]=0
    tw.loc[lreg_short > lreg_long,"signal"] = 1.0
    tw.loc[lreg_short < lreg_long,"signal"] =0.# -1.0
    #tw[(-2 < lreg ) & (lreg < 2)] = 0.0
    tw["returns"]=data.close.pct_change().shift(-1)
    return  tw
```

```python
def vhmm_trade(d, win_short=50, win_long=208,name='fischer_trade'):
    #p=(data.high+data.low)/2
    cross=np.log(pd.rolling_mean(d.close,win_long)) - np.log(pd.rolling_mean(d.close,win_short))
    thres=np.sign(cross)*pd.rolling_min(np.abs(cross),win_short)
    tw = pd.DataFrame(index=d.index.copy(),columns=["signal"])
    #print tw.head()
    tw.loc[:,"signal"]=0
    tw.loc[cross>0,"signal"] = 0.#-1.0
    tw.loc[cross<0,'signal'] = 1.0
    tw.loc[thres==cross,'signal']=0.
    #tw["price"]=d.price
    tw["returns"]=d.close.pct_change().shift(-1)# we lag one period as execution is laggy thu we take next close /!\
    return tw
```

```python
def volume_imbal(z):
    p=z.price.resample("5min",how="ohlc")
    p["vol"]=z.quantity.resample("5Min",how="sum")
    p["vol_buy"]=z[z.side==1].quantity.resample("5Min",how="sum")
    p["vol_sell"]=z[z.side==2].quantity.resample("5Min",how="sum")
    p["vol_ratio"]= p.vol_buy/p.vol_sell
    tw = pd.DataFrame(index=p.index.copy(),columns=["signal"])
    tw.loc[:,"signal"]=0
    tw.loc[p.vol_sell > p.vol_buy,"signal"] = 0.#-1.0
    tw.loc[p.vol_sell < p.vol_buy,'signal'] = 1.0
    #tw["smooth"]=pd.rolling_mean(tw.signal,3)
    #tw.loc[thres==cross,'signal']=0.
    #tw["price"]=p.close
    tw["returns"]=p.close.pct_change().shift(-1)
    #tw["vol"]=p.vol
    return tw
```

```python
def macd_str(df2, fast=21,slow=20,sig=9):
    macd=MACD(df2, fast,slow)
    tw=pd.DataFrame(index=macd.index)
    tw["signal"]=0
    tw["macd_cross"]=0
    tw.loc[macd.macdsignal<macd.macd,"macd_cross"]=1.
    tw.loc[macd.macdsignal>macd.macd,"macd_cross"]=0.
    tw.macd_cross=tw.macd_cross.diff()
    tw.loc[(tw.macd_cross == 1) & (macd.macdhist>macd.macd),"signal"]=1
    tw.loc[(tw.macd_cross == -1) & (macd.macdhist<macd.macd),"signal"]=0#-1
    tw.loc[tw.signal==0,'signal']=np.nan
    tw['signal']=tw.signal.fillna(method="ffill").fillna(0)
    tw["returns"]=df2.close.pct_change().shift(-1)
    return tw[["signal","returns"]]
```

```python
#print sigz.head()
#print sigz.index.shift(-1).head()

def filled_position(s):
    s=s[s.signal.diff()!=0]
    o=pd.DataFrame(index=s.index)
    o['opened']=o.index
    o['closed']=o.opened.shift(-1)
    o["side"]=s.signal
    idx=[]
    data=[]
    for k, row in o.iterrows():
    
        try:
            ret=d.loc[k:row.closed].close.pct_change().sum()
            data.append({"closed":row.closed, 'duration': row.closed-k,
                         'returns':ret, "side":row.side})
            idx.append(k)
        except:
            pass
    o=pd.DataFrame(data,index=idx)
    return o[["duration","returns","side"]]
```

```python
def stopLoss(d, o, thres=0.005):
    txn=[]
    idx=[]
    for k,r in o.iterrows():
        t=d[r.name:r.name+r.duration].returns.cumsum()*r.side
        ch_max=t.max()*r.side
        ch_min=t.min()*r.side
        #print k
        try:
            stop=t[t<-thres].index[0]
             
            tx={"closed":stop, 'duration': stop-k,
                         'returns':t[stop], "side":r.side, "otype":'STOP',
                         'max_chg':(t[:stop]*r.side).max(),
                         'min_chg':(t[:stop]*r.side).min()}
        except IndexError,e:
            #print 'zzzzz'
            tx={"closed":k+r.duration, 'duration': r.duration,
                         'returns':r.returns, "side":r.side,"otype":"LIMIT",
                         'max_chg':ch_max*r.side,'min_chg':ch_min*r.side}
        txn.append(tx)
        idx.append(k)
    o=pd.DataFrame(txn,index=idx)
    return o[["duration","returns","side","otype", 'max_chg','min_chg']]


def takeProfit(d, o, thres=0.05):
    txn=[]
    idx=[]
    for k,r in o.iterrows():
        t=d[r.name:r.name+r.duration].returns.cumsum()*r.side
        ch_max=t.max()*r.side
        ch_min=t.min()*r.side
        #print k
        try:
            stop=t[t>thres].index[0]
             
            tx={"closed":stop, 'duration': stop-k,
                         'returns':t[stop], "side":r.side, "otype":'TAKE_PROF',
                         'max_chg':(t[:stop]*r.side).max(),
                         'min_chg':(t[:stop]*r.side).min()}
        except IndexError,e:
            #print 'zzzzz'
            tx={"closed":k+r.duration, 'duration': r.duration,
                         'returns':r.returns, "side":r.side,"otype":r.otype,
                         'max_chg':ch_max*r.side,'min_chg':ch_min*r.side}
        txn.append(tx)
        idx.append(k)
    o=pd.DataFrame(txn,index=idx)
    return o[["duration","returns","side","otype", 'max_chg','min_chg']]
#o1.duration[0]
#orders[orders.returns < (orders.returns.mean() - 2*orders.returns.std())]
```
